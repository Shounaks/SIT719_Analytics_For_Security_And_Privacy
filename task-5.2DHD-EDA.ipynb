```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dataFrame = pd.read_csv('./data/Week_5_NSL-KDD-Dataset/kdd_merged.csv')
```

```python
column = 'service'

# Calculate value counts and sort by frequency
value_counts = dataFrame[column].value_counts().sort_values(ascending=False)

# Select top 5 values
top_5_values = value_counts.head(5)

# Create a bar plot using seaborn with top 5 values
plt.figure(figsize=(8, 6))
hist = sns.barplot(x=top_5_values.index, y=top_5_values.values)
plt.title(f"Distribution of {column} (Top 5 out of {len(value_counts)})")
plt.xlabel(column)
plt.ylabel("Frequency")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels if needed
plt.show()
```
