```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris = pd.read_csv(url, names=['sepal_length','sepal_width','petal_length','petal_width','class'])
iris.head()
```

#### Step #4: Write a code to check if there are any missing values in the dataframe

```python
iris.isnull().sum()
```

Step 5: Write a code to set the values of the rows 10 to 29 of the column 'petal_length' to NaN

```python
# iris[10:30]['petal_length'] = 'NaN' <- This will not work :)
iris.loc[10:29,'petal_length'] = np.nan
iris.head(31)
```

Step 6. Now again, check if there is any missing values (NaN) in the dataframe? Count, how many missing values.

```python
iris.isnull().sum()
```

Step 7. Substitute the NaN values to 10.0

This can be done with iris.loc(10:29,'petal_length') = 10.0 but the problem with this is that it will not remove all the NAN items. hence lets do it using methods which will fix the WHOLE Dataset


```python
iris['petal_length'].fillna(10.0, inplace=True)
iris.head(30)
```
